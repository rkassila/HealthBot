```python
import pandas as pd
import os
```

```python
path = "databases/diseases/dataset.csv"
diseases_df = pd.read_csv(os.path.join("..", path))
```

```python
diseases_df
```

```python
input_symptoms = ['itching', 'skin_rash']
```

```python
def check_symptoms(row, symptoms):
    row_symptoms = row.dropna().str.strip().tolist()  # Drop NaN values, strip whitespace, and convert to list
    return all(symptom in row_symptoms for symptom in symptoms)
```

```python
matched_diseases = diseases_df[diseases_df.apply(check_symptoms, symptoms=input_symptoms, axis=1)]
```
